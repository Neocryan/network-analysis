```python
try:
    sc.close()
except:
    pass
import re
from nltk.corpus import stopwords
import nltk
nltk.download('stopwords')
stopWords = set(stopwords.words('english'))
from pyspark import SparkConf, SparkContext
from pprint import pprint
from urllib.parse import unquote
sc = SparkContext()
pagecounts = sc.textFile('pagecounts-20160101-000000')
def bb1(x):
    print("id\t\thits\t\tsize\t\ttitle")
    for row in x:
        id, title, hits, size = row.split(' ')
        title = unquote(title)
        print("{}\t\t{}\t\t{}\t\t{}".format(id , hits, size,title))

def bb2(x):

    for row in x:
        id, title, hits, size = row.split(' ')
        title = unquote(title)
        print('''id:\t{}\n\nhits:\t{}\n\nsize:\t{}\n\ntitle\t{}'''.format(id , hits, size,title))
        print('-'*10 + '\n')
```

### 1. Retrieve the first k records and beautify

```python
k = 15
bb(pagecounts.take(k))
```

### 2.Determine the number of records the dataset has in total

```python
"Number of records: ", pagecounts.count()
```

### 3. Compute the min, max, and average page size.

```python
size = pagecounts.map(lambda x: int(x.split(' ')[3]))
# print("Min:\t", size.min(), "\nMax:\t", size.max(), "\nAvg:\t", size.mean())
print('Min:\t{} \nMax:\t{} \nMean:\t{}'.format(size.min(),size.max(),size.mean()))
```

### 4. Determine the record with the largest page size.

```python
size_record = pagecounts.map(lambda x: (int(x.split(' ')[3]), x))
max1 = size_record.max()[0]
max_record = size_record.filter(lambda x: max1 in x).map(lambda x: x[1])
bb2(max_record.collect())
```

### 5. Determine the record with the largest page size.

```python
size_record = pagecounts.map(lambda x: (int(x.split(' ')[3]), x))
max1 = size_record.max()[0]
max_size = size_record.filter(lambda x: max1 in x).map(lambda x: (int(x[1].split(' ')[2]), x[1]))
max2 = max_size.max()[0]
max_hit = max_size.filter(lambda x: max2 in x).map(lambda x: x[1])
bb2(max_hit.collect())
```

### 6. Determine the record with the longest page title.

```python
len_record = pagecounts.map(lambda x: (len(x.split(' ')[1]), x))
maxlen = len_record.max()[0]
max_len = len_record.filter(lambda x: x[0] == maxlen).map(lambda x: x[1])

print('there are total {} records that have the longest lenth of title'.format(len(max_len.collect())))
print('here just print 5 of them:\n')
bb2(max_len.take(5))
```

### 7. Use the results of question 3, and create a new RDD with the records that have greater page size from the average.

```python
average = size.mean()
size_record = pagecounts.map(lambda x: (int(x.split(' ')[3]), x))
records_greater = size_record.filter(lambda x: x[0] > average).map(lambda x: x[1])

print("Number of records:\t", records_greater.count())
size = records_greater.map(lambda x: int(x.split(' ')[3]))
# print("Min:\t", size.min(), "\nMax:\t", size.max(), "\nAvg:\t", size.mean())
print('NewMin:\t {} \nNewMax:\t {} \nNewMean: {}'.format(size.min(),size.max(),size.mean()))
```

### 8. Pageviews per project: compute the total number of pageviews for each project (as the schema shows, the first field of each record contains the project code).

```python
code_pageviews = pagecounts.map(lambda x: (x.split(' ')[0], int(x.split(' ')[2])))
result = code_pageviews.reduceByKey(lambda x,y : x+y).take(10)
print("code\t\ttotal_hits")
for (code, views) in result:
    print("{}\t\t{}".format(code, str(views)))
```

### 9. Report the 10 most popular pageviews of all projects , sorted by the total number of hits.

```python
result = code_pageviews.reduceByKey(lambda x,y : x+y).map(lambda x: (x[1], x[0])).sortByKey(False).take(10)

print("code\t\ttotal_hits")
for (code, views) in result:
    print("{}\t\t{}".format(code, str(views)))
```

### 10. Determine the number of page titles that start with the article ”The”.
How many of those page titles are not part of the English project

```python
start_code = pagecounts.map(lambda x: (x.split(' ')[1][0:3], x.split(' ')[0])).filter(lambda x: x[0] == 'The')
print('Number of Titles start with "The": ', start_code.count())
print('Number not part of "en": ', start_code.filter(lambda x: x[1] != 'en').count())
```

### 11. Determine the percentage of pages that have only received a single page view in this one hour of log data.

```python
nb_page1view = pagecounts.map(lambda x: int(x.split(' ')[2])).filter(lambda x: x == 1).count()
print("Sigle page percentage: {}%".format(round((nb_page1view / pagecounts.count() * 100),4)))
```

### 12. Determine the number of unique terms appearing in the page titles.

```python
c = pagecounts.flatMap(lambda x: [[q,1] for q in re.split(':|_|-|\.',("".join([z for z in x.split(' ')[1].\
                                                                               replace('/','')\
                                                                               .replace(' ','').lower().replace('%','')\
                                                                             ]))) if q not in stopWords])
term = c.reduceByKey(lambda a,b:a+b)
term.filter(lambda x : x[-1] == 1).collect()
```

### 13. Determine the most frequently occurring page title term in this dataset.

```python
c = pagecounts.flatMap(lambda x: [[q,1] for q in re.split(':|_|-|\.',("".join([z for z in x.split(' ')[1].\
                                                                               replace('/','')\
                                                                               .replace('\t','').lower().replace('%','')\
                                                                             ]))) if q not in stopWords])
topn = c.reduceByKey(lambda a,b:a+b).top(20,key = lambda x:x[-1])
print('the top 20 most frequently occurring page title term is:')
print('term\t\tcount')
for i in topn:
    print('{}\t{}'.format(i[0],i[1]))
    
```
